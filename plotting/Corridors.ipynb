```python
import matplotlib as mpl
mpl.rcParams['mathtext.fontset'] = 'cm'
# make sure to not use Level-3 fonts
mpl.rcParams['pdf.fonttype'] = 42

from plot_env import plot_env
import json
from plot_env import plot_env, plot_env_options
from plot_trajectory import plot_trajectory, plot_nodes, plot_trajectory_options
from color import get_color, get_colors
import matplotlib.pyplot as plt
```

## Corridor Radii

```python
data = json.load(open('../bin/log/corridor_radii.json', 'r'))
seeds = 5  # how many seeds where generated per corridor radius
radii = len(data["runs"]) // seeds
print('There are %i radius configurations' % radii)
plt.figure(figsize=(radii * 5, seeds * 5))
for i, run in enumerate(data["runs"]):
    plt.subplot(seeds, radii, i+1)
    plot_env(run["environment"], draw_start_goal_thetas=False, draw_start_goal=False, set_title=False)
    plt.title("$r = %.1f$ (seed: %i)" % (run["settings"]["env"]["grid"]["corridor"]["radius"], run["settings"]["env"]["grid"]["seed"]), fontsize=15)
plt.savefig("corridor_radii.pdf", bbox_inches='tight')
plt.savefig("corridor_radii.png", bbox_inches='tight')
```

## Obstacle Ratios

```python
data = json.load(open('../bin/log/obstacle_ratios.json', 'r'))
seeds = 5  # how many seeds where generated per corridor radius
ratios = len(data["runs"]) // seeds
print('There are %i obstacle ratio configurations' % ratios)
plt.figure(figsize=(ratios * 5, seeds * 5))
for i, run in enumerate(data["runs"]):
    plt.subplot(seeds, ratios, i+1)
    plot_env(run["environment"], draw_start_goal_thetas=False, draw_start_goal=False, set_title=False)
    plt.title("$r = %.2f$ (seed: %i)" % (run["settings"]["env"]["grid"]["random"]["obstacle_ratio"], run["settings"]["env"]["grid"]["seed"]), fontsize=15)
plt.savefig("obstacle_ratios.pdf", bbox_inches='tight')
plt.savefig("obstacle_ratios.png", bbox_inches='tight')
```
